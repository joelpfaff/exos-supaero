```python
import urllib.request
import gzip
```

```python
localfile = 'pageviews-20190331-120000'
```

```python
url = 'https://dumps.wikimedia.org/other/pageviews/2019/2019-03/pageviews-20190331-120000.gz'

r = urllib.request.urlopen(url)
s_out = gzip.decompress(r.read())

with open(localfile, 'wb') as f:  
    f.write(s_out)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Prepare dataset
import pandas as pd
df = pd.read_csv(localfile, sep=' ', header=None)
df.columns = ['code', 'title', 'nbrequest', 'total']
```
